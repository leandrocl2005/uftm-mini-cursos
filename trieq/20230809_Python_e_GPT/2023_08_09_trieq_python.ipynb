```python
# chat GPT 3.5
# Problema 1: Faça um programa mais simples possível em Python para calcular
# a média de dois números inteiros digitados pelo usuário.

num1 = int(input("Digite o primeiro número inteiro: "))
num2 = int(input("Digite o segundo número inteiro: "))
media = (num1 + num2) / 2
print(f"A média dos números {num1} e {num2} é: {media}")
```

```python
# chat GPT 3.5
# Problema 2: Faça um programa o mais simples possível em Python para calcular a média de três números quaisquer digitados pelo usuário. O usuário irá digitar os três números
# numa só linha separados por espaço.

numeros = input("Digite os três números separados por espaço: ")
num1, num2, num3 = map(int, numeros.split())
media = (num1 + num2 + num3) / 3
print(f"A média dos números é: {media}")
```

```python
# chat GPT 3.5
# Problema 2 (Correção): No meu último pedido, eu falei que os 3 números digitados pelo usuário eram quaisquer, isso quer dizer que os 3 números digitados pelo usuário podem ser inteiros ou pontos
# flutuantes. Corrija o programa.

numeros = input("Digite os três números separados por espaço: ")
num1, num2, num3 = map(float, numeros.split())
media = (num1 + num2 + num3) / 3
print(f"A média dos números é: {media}")
```

```python
# chat GPT 3.5
# Problema 3: Faça um programa o mais simples possível em Python para calcular a
# média dos números de um arquivo de nome "numeros.txt". No arquivo, cada
# número encontra-se numa linha diferente. Os números podem ser pontos
# flutuantes ou inteiros.

# Ele fez um programa que trata o erro da não existência do arquivo.

# Problema 3 (complemento): Seu programa ficou muito complicado, faça um mais
# simples. Não precisa tratar exceções.

with open("numeros.txt") as arquivo:
    numeros = arquivo.readlines()

numeros = [float(numero.strip()) for numero in numeros]
media = sum(numeros) / len(numeros)
print(f"A média dos números é: {media}")
```

```python
# chat GPT 3.5

# Problema 4: Faça um programa o mais simples possível em Python para mostrar um
# gráfico de linha dos números de um arquivo de nome "numeros.txt". No arquivo,
# cada número encontra-se numa linha diferente. Os números podem ser pontos
# flutuantes ou inteiros.

import matplotlib.pyplot as plt

with open("numeros.txt") as arquivo:
    numeros = arquivo.readlines()

numeros = [float(numero.strip()) for numero in numeros]

plt.plot(numeros)
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('Gráfico de Linha dos Números')
plt.show()
```

```python
# chat GPT 3.5

# Problema 5: Acrescente ao gráfico anterior a média móvel considerando os 3
# valores anteriores ao índice atual.


import matplotlib.pyplot as plt
import pandas as pd

with open("numeros.txt") as arquivo:
    numeros = arquivo.readlines()

numeros = [float(numero.strip()) for numero in numeros]

# Calcula a média móvel considerando os 3 valores anteriores ao índice atual
media_movel = pd.Series(numeros).rolling(window=4).mean()

plt.plot(numeros, label='Números')
plt.plot(media_movel, label='Média Móvel')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('Gráfico de Linha dos Números com Média Móvel')
plt.legend()
plt.show()
```

```python
# gpt 3.5

# Problema 8: Não gostei da média móvel. Faça um novo gráfico, no lugar da média
# móvel dos últimos 3 dias para estimar o índice atual, faça um código em Python
# simples que usa estimativas do modelo Prophet para séries temporais.

import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
from prophet import Prophet

# Obtém os dados do Yahoo Finance para as ações do Google nos últimos 4 anos
ticker = "GOOGL"
data = yf.download(ticker, period="4y")["Close"]

# Criar um DataFrame com as datas e valores de fechamento
df = pd.DataFrame({"ds": data.index, "y": data.values})

# Criar e treinar o modelo Prophet
modelo = Prophet()
modelo.fit(df)

# Criar um DataFrame com as datas futuras para fazer a previsão dos próximos 365 dias
futuro = modelo.make_future_dataframe(periods=365)

# Fazer a previsão dos valores de fechamento para os próximos 365 dias
previsao = modelo.predict(futuro)

# Plota o gráfico de linha dos valores de fechamento e a previsão
fig = modelo.plot(previsao, xlabel='Data', ylabel='Valor de Fechamento')
plt.title(f'Valores de Fechamento das Ações do {ticker} nos Últimos 4 Anos com Previsão para os Próximos 365 Dias')
plt.show()
```
